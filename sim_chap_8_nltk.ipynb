```python
#sim_chap_8 : 영문 텍스트 분석하기


import nltk
nltk.download('stopwords')  # nltk 패키지를 사용하기 위한 데이터를 다운로드 하기 - 시간이 몇 분 걸림

#nltk 패키지가 정상적으로 설치 되었는 지 아래의 방법으로 테스트하기
from nltk.corpus import brown
# brown.words()

#불용어 제거를 위한 불용어 사전 로딩
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer

# 스티브잡스 연설문 분석
# 파일 불러와서 형태소 분석하기
data1 = open("/content/drive/MyDrive/인턴십공부/steve.txt", encoding='cp949').read() 


# word_tokenize 함수는 축약형을 제대로 분리할 수 없음
# 그래서 아래의 방법을 추천
from nltk.tokenize import WordPunctTokenizer

tokenizer = WordPunctTokenizer()
new_data2 = tokenizer.tokenize(data1)
print(new_data2)
```

```python
tokenizer = RegexpTokenizer("[\w']+")


new_data3 = tokenizer.tokenize(data1)
print(new_data3)

# 불용어 제거하기
data4 = [ each_word for each_word in new_data3 
          if each_word not in stopwords.words() ]
data4
```

```python
# 단어별로 언급빈도 집계하기
from collections import Counter

data5 = Counter(data4)

data6 = data5.most_common(100)

data7 = dict(data6)
```

```python
# 집계된 단어를 그래프로 표시하기
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
g_data4 = nltk.Text(data4 , name=" 다빈도 단어 그래프 출력하기")
g_data4.plot(20)
```
